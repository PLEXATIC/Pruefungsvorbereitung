```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol("x")
expression = sp.cos(x)
dy_dx = expression.diff(x)
print(dy_dx)

print(expression)
xs = np.arange(-1, 1, 0.01)
ys = [expression.subs(x, x_val) for x_val in xs]
abls = [dy_dx.subs(x, x_val) for x_val in xs]
plt.plot(xs, ys)
plt.plot(xs, abls)
```

```python
import sympy as sp

x = sp.Symbol("x")
inner_function = sp.Symbol("a")
inner_function = -(1/2)*(x**2) + (5/2)*x - 3
print(inner_function)
k = sp.Symbol("k")
outer_function = sp.Sum(inner_function**k, (k, 0, sp.oo))

print(outer_function)


```
