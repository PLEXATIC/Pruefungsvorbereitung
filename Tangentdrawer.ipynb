```python
import pygame as pg
import numpy as np
import math
width = 1920
height = 1080
pg.init()
screen = pg.display.set_mode((width, height))
background_color = (0, 0, 0)
line_color = (255, 255, 255)
tangent_color = (255, 0, 125)
scale_x = 12
scale_y = 0.04

balls = []
ball_vels = []
ball_velocities = 0.01

def f(x):
    x = x*scale_x
    return math.sin(x*math.pi)*(math.cos(2*math.pi)/2)*(x*math.sin(3*x))*scale_y

def derivative(f, x, dx):
    return (f(x+dx)-f(x))/dx

running = True
clock = pg.time.Clock()
while running:
    mouse_x, mouse_y = pg.mouse.get_pos()
    mouse_x /= width/2
    mouse_x -= 1

    for event in pg.event.get():
        if event.type == pg.QUIT:
            running = False
        if event.type == pg.MOUSEBUTTONDOWN:
            balls.append(mouse_x)
            ball_vels.append(0)
    
    screen.fill(background_color)
    points = []
    for x in np.arange(-1, 1, 0.001):
        y = f(x)
        point = (width*(x+1)/2, 0.5*height-(0.5*height*f(x)))
        points.append(point)

    for i in range(len(points)-1):
        x1, y1 = points[i]
        x2, y2 = points[i+1]
        pg.draw.line(screen, line_color, points[i], points[i+1])
    
    y = f(mouse_x)
    m = derivative(f, mouse_x, 0.0001)
    b = y-m*mouse_x
    for x in np.arange(mouse_x-0.3, mouse_x+0.3, 0.001):
        y = m*x + b
        cx = (x+1)*width/2
        cy = height/2-(y*height/2)
        pg.draw.circle(screen, tangent_color, (cx, cy), 1, 0)
    
    for index, bx in enumerate(balls):
        x = bx
        y = f(x)
        ball_vels[index] -= derivative(f, x, 0.0001)
        ball_vels[index] *= 0.999
        pg.draw.circle(screen, (0, 125, 255), ((x+1)*width/2, height/2-(y*height/2)), 10)
        balls[index] += ball_vels[index]*0.0001


    pg.display.flip()
    clock.tick(30)
pg.quit()

```
