## Funktionen 
Funktionen sind dadurch definiert, dass sie jedem x Wert genau einen y Wert zuweiesen.

Funktionen haben einen Definitionsbereich $D(f)$ und einen Wertebereich $W(f)$

Ist eine Funktion $F(x) = mx + b$ so ist $m$ die Steigung der Funktion


```python
import matplotlib.pyplot as plt
import numpy as np
import math

def f(x):
    m = 0.6
    b = 0
    return m*x + b

xs = np.arange(-1, 2, 0.05)
ys = [f(x) for x in xs]
plt.plot(xs, ys)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.show()
```

Die Heavysidefunktion ist $y = 0$ falls $x < a$ und $y = 1$ falls $x >= a$.
Hier der Plot f√ºr die Heavysidefunktion mit a = 0.5 von 0 bis 1:

```python
def h(x, a):
    if x-a < 0:
        return 0
    else:
        return 1

xs = np.arange(0, 1, 0.01)
ys = [h(x, 0.5) for x in xs]
plt.plot(xs, ys)
plt.show()
```

Um Funktionen eine verschobene Version einer Funktion zu erzeugen kann $g(x) = f(x-a)$ genutzt werden um die Funkion G zu erzeugen, die gleich wie F, aber verschoben ist:

```python
def f(x):
    return math.sin(x)

def g(x, a):
    return f(x-a)

xs = np.arange(-10, 10, 0.1)
ys_f = [f(x) for x in xs]
plt.plot(xs, ys_f)
for a in np.arange(-1, 1, 0.5):
    ys_g = [g(x, a) for x in xs]
    plt.plot(xs, ys_g)
plt.ylim([-1, 1])
plt.xlim([-1, 1])
plt.show()
```

Eine Zusammengesetzte Funktion aus $f$ und $g$ w√§re beispielsweise ein √ñlteppich, der sich √ºber Zeit ausbreitet.
Man vereinfacht, so dass $A = f(g(x))$ die Fl√§che des √ñlteppiches ist. 
Dabei ist $f(x) = 2*x*pi$ und $g(x) = 1+x$

```python
def f(x):
    return 2*x*math.pi
def g(x):
    return 1+x
def A(x):
    return f(g(x))

ts = np.arange(0, 10, 0.1)
a_vals = [A(t) for t in ts]
plt.plot(ts, a_vals)
plt.show()
```

Polynome haben die Eigenschaft, dass man sie in in der form $y = (x+a)(x+b)...$ schreiben, wobei jeweils die nulstellen bei -a, -b etc. liegen.

```python
def f(x):
    return (x+1)*(x+2)*(x-1)*(x-2)

xs = np.arange(-5, 5, 0.02)
ys = [f(x) for x in xs]
plt.plot(xs, ys)
plt.grid()
plt.show()
```

## Arithmetische Folge
Die Arithmetische Folge ist eine Folge die zus√§tzlich durch 
$a_{n+1} = a_n + d$
definiert. 

Kennt man das Startglied $a_0$ sowie die Differenz $d$, so kann man jedes Element $a_n$ mit $a_n = a_0 + n*d$ berechnen:


```python
ari = [1, 4, 7, 10]
d = 3
a_0 = 1
a_2 = a_0 + 2*d
print(a_2)
```

Die Summe von arithmetischen Gliedern kann durch die Formel ${n}\frac{a_1 + a_n}{2}$ berechnet werden. (hier ist n der Index des letzten Elementes und die Folge beginnt bei $a_1$)

```python
def ari_sum(ari):
    n = len(ari)
    a_1 = ari[0]
    a_n = ari[n-1]
    return n*(a_1+a_n)/2

print(ari_sum(ari))
```
